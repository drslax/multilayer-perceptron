<h1>The perceptron</h1>
<br>
Invented in 1957 by Frank Rosenblatt at the Cornell Aeronautical Laboratory, a perceptron is the simplest neural network possible: a computational model of a single neuron. A perceptron consists of one or more inputs, a processor, and a single output.
<img src="img/per_bias.png">

A perceptron follows the “feed-forward” model, meaning inputs are sent into the neuron, are processed, and result in an output. In the diagram above, this means the network (one neuron) reads from left to right: inputs come in, output goes out.

<h6>Now we will create a Perceptron Class</h6>

```python
import random
import numpy as np
from matplotlib import pyplot as plt
import time

class Perceptron():
    def __init__(self, n):
        self.n = n
        self.weight = []
        for i in range(self.n):
            self.weight.append(random.uniform(-1,1))

    def sign(self, x):
         if x >= 0:
            return 1
         return -1
    def feedforward(self, inp):
        sumu = 0
        for i in range(self.n):
            sumu += inp[i] * self.weight[i]
        return self.sign(sumu)
    
    def train(self, inputs, desired_out):
        lr = 0.1
        guess = self.feedforward(inputs)
        error = desired_out - guess
        for i in range(self.n):
            self.weight[i] += error * inputs[i] * lr
```

```python
class Trainer():
    def __init__(self, x, y, a):
        self.inputs = [x, y, 1]
        self.answer = a
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
fig.show()
fig.canvas.draw()
n = 1000
xy = [[random.randint(-10,10), random.randint(-10,10)] for i in range(n)]

def f(var):

    y_line = 2 * var[0] + 1
    if var[1] > y_line:
        return(1)
    else:
        return(-1)

p = Perceptron(3)
ts = [Trainer(xy[i][0], xy[i][1], f(xy[i])) for i in range(n)]

for t in ts:
    p.train(t.inputs, t.answer)
    x = np.linspace(-6,6,2)
    y = 2*x+1
    plt.plot(x, y, '-g')
    if (p.feedforward(t.inputs) == 1):
        ax.plot(t.inputs[0], t.inputs[1],'ko',markersize=1, color='red')
    else:
        ax.plot(t.inputs[0], t.inputs[1],'ko',markersize=1, color='blue')
    time.sleep(0)
```
